```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
Kholi = [500,700,1100,1500,1800,1200,1700,1300,900,1500]
Sehwag = [300,7100,1100,1200,1300,1210,1440,1200,1100,500]
plt.plot(years, Kholi, 'ro--',  label = "Virat Kholi")
plt.plot(years, Sehwag, 'g^:',  label = "Virendra Sehwag")
plt.xlabel("Years")
plt.ylabel("Runs")
plt.title("Performace Comparison")
plt.legend()
```
