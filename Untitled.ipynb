```python
import matplotlib.pyplot as plt
```

```python
study_hours = [1,2,3,4,5,6,7,8,9]
exam_scores = [40,45,50,55,60,65,75,85,90]
```

```python
#colors = ['red' if score<50 else 'green' for score in exam_scores]
sizes = [score *2 for score in exam_scores]

plt.scatter(study_hours, exam_scores, c=exam_scores, cmap='viridis')
plt.grid(True)
plt.colorbar(label="Score")
plt.xlabel("Study Hours")
plt.ylabel("Exam Scores")
plt.title("Study vs Marks")
```

```python
x = [1,2,3,4,5]
y1 = [i * 2 for i in x]
y2 = [i ** 2 for i in x]

y3 = [i ** 0.5 for i in x]
y4 = [10-i for i in x]

plt.figure(figsize=(8,6))

plt.subplot(2,2,1)
plt.plot(x,y2)
plt.title('x*2')

plt.subplot(2,2,2)
plt.plot(x,y1)
plt.title('x Squared')

plt.subplot(2,2,3)
plt.plot(x,y3)
plt.title('sqrt(x)')

plt.subplot(2,2,4)
plt.plot(x,y4)
plt.title('10-x')

plt.tight_layout()
plt.show()
```
